# [dbo].[storedListar]

```sql
/****** Object:  StoredProcedure [dbo].[storedListar]    Script Date: 31/5/2025 11:48:49 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[storedListar] as
SELECT A.Id, A.Codigo, A.Nombre, A.Descripcion, C.Descripcion AS Tipo, M.Descripcion AS Marcas, A.Precio, A.ImagenUrl, A.IdCategoria, A.IdMarca 
FROM ARTICULOS A, CATEGORIAS C, MARCAS M 
WHERE A.IdCategoria = C.Id AND A.IdMarca = M.Id 
GO
```
