# [dbo].[storedAltaArticulo]

```sql
/****** Object:  StoredProcedure [dbo].[storedAltaArticulo]    Script Date: 8/8/2025 19:58:30 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE  [dbo].[storedAltaArticulo]
    @Id INT = 0,
    @Codigo NVARCHAR(50),
    @Nombre NVARCHAR(100),
    @Descripcion NVARCHAR(200),
    @Precio DECIMAL(18,2),
    @PrecioDescuento DECIMAL(18,2),
    @ImagenUrl NVARCHAR(200),
    @IdCategoria INT
AS
BEGIN
    IF @Id = 0
    BEGIN
        INSERT INTO Articulos (Codigo, Nombre, Descripcion, Precio, PrecioDescuento, ImagenUrl, IdCategoria)
        VALUES (@Codigo, @Nombre, @Descripcion, @Precio, @PrecioDescuento, @ImagenUrl, @IdCategoria)
    END
    ELSE
    BEGIN
        UPDATE Articulos
        SET Codigo = @Codigo,
            Nombre = @Nombre,
            Descripcion = @Descripcion,
            Precio = @Precio,
            PrecioDescuento = @PrecioDescuento,
            ImagenUrl = @ImagenUrl,
            IdCategoria = @IdCategoria
        WHERE Id = @Id
    END
END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[storedListar]    Script Date: 8/8/2025 19:58:30 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[storedListar]
AS
SELECT A.Id, A.Codigo, A.Nombre, A.Descripcion,
       C.Descripcion AS Tipo,
       M.Descripcion AS Marcas,
       A.Precio, A.PrecioDescuento,
       A.ImagenUrl, A.IdCategoria, A.IdMarca,
       A.FechaCreacion
FROM ARTICULOS A
INNER JOIN CATEGORIAS C ON A.IdCategoria = C.Id
INNER JOIN MARCAS M ON A.IdMarca = M.Id;
GO
```
